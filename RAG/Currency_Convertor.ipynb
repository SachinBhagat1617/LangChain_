```python
import os
import requests
from langchain_core.tools import InjectedToolArg,tool
from langchain_huggingface import HuggingFaceEndpoint,ChatHuggingFace
from dotenv import load_dotenv
from typing import Annotated
load_dotenv(override=True)

api_key = os.getenv("EXCHANGE_RATE_API_KEY")

@tool
def get_conversion_factor(base_cuurency: str, target_currency: str)-> float:
    """Get conversion factor between two currencies."""
    url=f"https://v6.exchangerate-api.com/v6/{api_key}/pair/{base_cuurency}/{target_currency}"
    response=requests.get(url)
    return response

@tool
def convert_currency(amount: float, conversion_factor: float) -> float:
    """Convert amount from one currency to another using conversion factor."""
    return amount * conversion_factor

```

```python
convert_currency.args
```

```python
response = get_conversion_factor.invoke({"base_cuurency": "USD", "target_currency": "INR"})
conversion_factor = response.json()['conversion_rate']
print(f"Conversion Factor from USD to INR: {conversion_factor}")
```

```python
llm=HuggingFaceEndpoint(
    repo_id="mistralai/Mistral-7B-Instruct-v0.2",
    task="text-generation",
)
model=ChatHuggingFace(llm=llm)

llm_with_tools = model.with_tools([get_conversion_factor, convert_currency])
```
